```python
from pathlib import Path
from collections import defaultdict

OUTPUT_DIR = Path("/local/scratch/carlyn.1/dna/training_output/")
EXP_NAME = "deepnet_7_17"

exp_path = OUTPUT_DIR / EXP_NAME

count = 0
chromosome_counts = defaultdict(int)
for root, dirs, files in exp_path.walk():
    for file in files:
        if file == "model.pt":
            count += 1
            for i in range(1, 22):
                
                if str(root).split("_")[-1] == str(i):
                    chromosome_counts[i] += 1
total = 4 * 2 * 2 * 21

print(f"{count / total:.2%} of models trained")
print("Chromosome counts:")
for i in range(1, 22):
    print(f"Chromosome {i}: {chromosome_counts[i]} models")
```
