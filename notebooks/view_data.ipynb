```python
# What is the range of RMSE loss?
import os
import numpy as np

def isnum(x):
    return x.replace(".", "").isdigit()

def print_stats(x):
    r = lambda x: round(x, 4)
    print(f"Min: {r(x.min())} | Max: {r(x.max())}")
    print(f"Mean: {r(x.mean())} | STD: {r(x.std())}")

vals = []
for root, dirs, paths in os.walk("../results/chromosome"):
    for p in paths:
        if p == "out.log":
            full_path = os.path.join(root, p)
            with open(full_path, 'r') as f:
                lines = f.readlines()
                #print(lines[-3].strip())
                parts = lines[-3].strip().split(" ")
                if len(parts) < 10: continue
                train = parts[2]
                val = parts[6]
                test = parts[10]
                if not isnum(train) or not isnum(val) or not isnum(test):
                    continue
                vals.append([float(train), float(val), float(test)])
vals = np.array(vals)

print("Training Statistics")
print_stats(vals[:, 0])
print("Validation Statistics")
print_stats(vals[:, 1])
print("Testing Statistics")
print_stats(vals[:, 2])
        
```
